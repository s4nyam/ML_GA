Task1: **Given a randomly generated bit sting of length 100, create a genetic algorithm that through mutation tries to obtain a string containing only 1s, i.e., 111111...111111.**



# Bit Flip Mutation

```python
# Just using bit flip - No Emergence or Evolution

import random
import numpy as np
length = 100
string = ""
for each_bit in range(length):
  generated_bit = str(random.randint(0,1))
  string = str(string) + str(generated_bit)
print("Given Random string of length "+ str(length) + " is: " + string)
goal_string = "1111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111"
print("Given Evolve string of length 100 is: "+ goal_string)

# Difference or Distance using Bit Flip Mutation

difference_counter = 0
randomstring = np.array(list(string))
goalstring = np.array(list(goal_string))
updated_string = ""
for i in range(length):
    difference_for_bit = int(randomstring[i]) - int(goalstring[i])
    if(difference_for_bit == 0):
      pass
    else:
      difference_counter = difference_counter + 1
      randomstring[i] = '1'
      mutated_string = randomstring
      final_a = ""
      for e in mutated_string:
        final_a = final_a + e
      print("Evolved New string is: " + final_a)
      print(difference_counter)



```

```python
length = 100
def produce_mutations_without_fitness(parent1,parent2):
  crossover_point = random.randint(0, 99)
  print("Randomly generated Crossover point :", crossover_point)
  parent1_array = np.array(list(parent1))
  parent2_array = np.array(list(parent2))
  for i in range(crossover_point, length):
    temp_value = parent1_array[i]
    parent1_array[i] = parent2_array[i]
    parent2_array[i] = temp_value
  parent1_array_string = ''.join(parent1_array)
  print("parent 1 mutated string: "+parent1_array_string)
  
  parent2_array_string = ''.join(parent2_array)
  print("parent 2 mutated string: "+parent2_array_string)
  parent1 = parent1_array_string
  parent2 = parent2_array_string



# Two random parents of 100 size
parent1 = ""
for each_bit in range(length):
  generated_bit = str(random.randint(0,1))
  parent1 = str(parent1) + str(generated_bit)
# print("Given Random parent1 of length "+ str(length) + " is: " + parent1)

parent2 = ""
for each_bit in range(length):
  generated_bit = str(random.randint(0,1))
  parent2 = str(parent2) + str(generated_bit)
# print("Given Random parent2 of length "+ str(length) + " is: " + parent2)
print("parent 1 we started with: ",parent1)
print("parent 2 we started with: ", parent2)

# Genearting for 5 generations
for i in range(5):
  print("---------------------------------------------------------------------------")
  print('Generation '+ str(i) + ' starts: ')
  produce_mutations_without_fitness(parent1,parent2)
  print("---------------------------------------------------------------------------")
```
